# **Data Clearnsing, Data Tranform**

```python
results=pd.read_csv("satisfaction_SU.csv") #นำข้อมูลเข้า
```

```python
result_df= pd.DataFrame(results)
result_df.shape #ตรวจสอบแถวและคอลัมน์
```

```python
result_df.columns #คอลัมน์มีชื่ออ่ะไรบ้าง
```

```python
result_df.info() #ตรวจสอบว่าในแต่ล่ะ คอลัมน์ มีกี่แถว และ เป็นข้อมูลอ่ะไร
```

```python
result_df.head()  #ลองแสดงผล ชุด data
```

```python
result_df.isnull().sum() #ตรวจสอบข้อมูล ที่เป็น null
```

```python
result_df.rename(columns={'คณะ/ หน่วยงาน ':'คณะ'},inplace=True)
result_df.rename(columns={'การตอบคำถามที่นักศึกษาสงสัย ผ่านทางช่องทางออนไลน์ {เช่น กยศ/กรอ}':'การตอบคำถามที่นักศึกษาสงสัยออนไลน์'},inplace=True)
result_df.rename(columns={'การให้บริการสะดวกรวดเร็ว ครบถ้วน ถูกต้อง ในรูปแบบออนไลน์':'การให้บริการทางออนไลน์ดี'},inplace=True)
result_df.rename(columns={'การประชาสัมพันธ์ เกี่ยวกับกระบวนการ/ขั้นตอน { การลดค่าเทอม หรือกิจกรรม เป็นต้น } มีขั้นตอนและระยะเวลาในการให้บริการชัดเจน':'การประชาสัมพันธ์ดี'},inplace=True)
result_df.rename(columns={'นักศึกษาอยากได้รับการช่วยเหลือสวัสดิการต่างๆ ในการสนับสนุนการเรียนออนไลน์อะไรบ้าง เช่น ซิมการ์ด , ทุนการศึกษา , อุปกรณ์คอมพิวเตอร์':'นักศึกษาอยากได้รับความช่วยเหลือด้านใด'},inplace=True)
#แปลงชื่อ คอลัมน์ เป็นชื่อสั้นๆเข้าใจง่าย
```
